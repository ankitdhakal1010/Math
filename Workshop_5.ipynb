```python
import numpy as np
import matplotlib.pyplot as plt
""" Problem 2: Create 2-dimensional position vectors and plot them. """
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
# Add vectors V and W to the plot
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, U[0], U[1], U[2], color='r', label='U')
ax.quiver(0, 0, 0, V[0], V[1], V[2], color='g', label='V')
ax.quiver(0, 0, 0, W[0], W[1], W[2], color='b', label='W')

# Set labels
ax.set_xlim([0, 1.2])
ax.set_ylim([0, 1.2])
ax.set_zlim([0, 1.2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
U = np.array([-0.5, 0.5])
V = np.array([1, 1])

# Add vectors
W = U + V  # Resultant vector

# Create plot
plt.figure()

# Plot vectors from origin
plt.quiver(0, 0, U[0], U[1], angles='xy', scale_units='xy', scale=1, color='r', label='U')
plt.quiver(0, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='g', label='V')
plt.quiver(0, 0, W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b', label='W = U + V')

# Set axis limits for better view
plt.xlim(-1, 2)
plt.ylim(-1, 2)

# Add labels and grid
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.legend()

# Show plot
plt.show()
```

```python
import numpy as np
def magnitude(vector):
  return np.sqrt(np.sum(vector**2))
U = np.array([1.5,1.5,2.0])
mag = magnitude(U)
print("magnitude of U is :",mag)
```

```python
import numpy as np
def direction_cosines(vector):
  magnitude = np.sqrt(np.sum(vector**2))
  return vector / magnitude
O = np.array([1.5,1.5,2.0])
dc = direction_cosines(O)
print("direction_cosines is:",dc)
```

```python
import numpy as np
def dot_product(a,b):
  return np.sum(a*b)
a = np.array([1.5,1.5,2.0])
b = np.array([-1.5,0.5,2.0])
result = dot_product(a,b)
print("dot product:",result)
```

```python
import numpy as np
def cross_product(a,b):
  return np.cross(a,b)
a = np.array([1.5,1.5,2.0])
b = np.array([-1.5,0.5,2.0])
result = cross_product(a,b)
print("cross product:",result)
```
