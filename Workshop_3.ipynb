9.Plot the trigonometric COSINE function in interval [−3𝜋,3𝜋], i.e., 𝑦 = cos (𝑥), −3𝜋 ≤ 𝑥 ≤3𝜋

```python
x = []
y=[]
start = -3*math.pi
end = 3*math.pi

total_points = 750

step =(end - start) / total_points

for i in range(total_points + 1):
    x_val = start + (i * step)
    x.append(x_val)
    y.append(math.cos(x_val))

plt.plot(x,y,marker='o')

plt.show()
```

11. Plot the above circle and the ellipse on the same axes. This is give an idea about the intersection points of the two conic sections.

```python
center = [2,2]

def circle():
    x=[]
    y=[]
    r=4
    center=[2,2]

    theta_start = 0.0
    theta_end = 2.5*math.pi

    step_size = 0.125
    steps = math.ceil((theta_end - theta_start)/step_size)

    theta = theta_start

    for i in range(0,steps):
        x.append(r*math.cos(theta)+center[0])
        y.append(r*math.sin(theta)+center[1])
        theta += step_size
    return x,y

def ellipse():
    x = []
    y= []

    a=3
    b=5
    center = [2,2]

    theta_start = 0.0
    theta_end=3*math.pi
    step_size = 0.125

    steps = math.ceil((theta_end-theta_start)/step_size)

    theta = theta_start

    for i in range(steps):
        x.append(a*math.cos(theta)+center[0])
        y.append(b*math.sin(theta)+center[1])
        theta += step_size
    return x,y


x1,y1 = circle()
x2,y2 = ellipse()

plt.figure()
plt.plot(center[0],center[1],'.')
plt.plot(x1,y1,'-')
plt.plot(x2,y2,'.')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

10. Plot a circle with center at (a, b) and radius r. The values of a=b=2, and r=4, using a while loop.

```python
x=[]
y=[]
r=4
center=[2,2]

theta_start = 0.0
theta_end = 2.5*math.pi

step_size = 0.125
steps = math.ceil((theta_end - theta_start)/step_size)

theta = theta_start

for i in range(0,steps):
    x.append(r*math.cos(theta)+center[0])
    y.append(r*math.sin(theta)+center[1])
    theta += step_size

plt.figure()

plt.plot(center[0],center[1],'.')
plt.plot(x,y,'-')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')

ax=plt.gca()
ax.set_aspect('equal',adjustable='box')
plt.show()
```
